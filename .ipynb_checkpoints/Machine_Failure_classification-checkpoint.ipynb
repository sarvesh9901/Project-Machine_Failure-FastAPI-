```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('machine_failure_cleaned.csv')
df.head()
```

```python
df = df.rename({'Rotational speed [rpm]':'Rotational_speed_rpm','Torque [Nm]':'Torque_Nm', 'Tool wear [min]':'Tool_wear_min','Machine failure':'Machine_failure'},axis=1)
```

```python
x= df.drop(columns=['Machine_failure'], axis=1)
y = df.Machine_failure
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Define classifiers and their parameter grids
classifiers = {
    'Logistic Regression': (LogisticRegression(), {
        'C': [0.001, 0.01, 0.1, 1, 10, 100],  
        'penalty': ['l1', 'l2'],  
        'solver': ['liblinear']
    }),
    'Random Forest': (RandomForestClassifier(), {
        'n_estimators': [100, 200, 300],  
        'max_depth': [None, 10, 20, 30],  
        'min_samples_split': [2, 5, 10],  
        'min_samples_leaf': [1, 2, 4]
    })
}

# Iterate over classifiers
for clf_name, (clf, param_grid) in classifiers.items():
    print(f"Tuning hyperparameters for {clf_name}")
    
    # Create GridSearchCV instance
    grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy')
    
    # Fit GridSearchCV
    grid_search.fit(x_train, y_train)
    
    print("Best parameters found:")
    print(grid_search.best_params_)
    print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
    print("\n")
```
